# Comaparision Operators, it will give boolean answers

```python
x = 5
y = 10
# perform comparision operator
print(x == y) # equal to
print(x != y) # not equal to
print(x > y) # greater than
print(x < y) # less than
print(x >= y) # greater than or equal to
print(x <= y) # less than or equal to
```

# Logical Operator

```python
x = 5
y = 10
z = 15
# perform logical operator
print(x < y and y < z) # and
print(x < y or x > z) # or
print(not(x < y)) # not
#
```

# Bitwise Operator

see how to convert binary to decimal and decimal to binary, bitwise give fast result

bitwise and, the decimal value will be first converted to binary and then the and operation will be performed and then the output will be converted into decimal again

```python
x = 10
y = 6
# perform all bitwise operations
print(x & y) # and
print(x | y) # or
print(x ^ y) # xor
print(~x) # not
print(x << 1) # left shift
print(x >> 1) # right shift
```

# Assignment Operator

```python
x = 20
y = 3
# perform all assignment operations
x += y # x = x + y
print(x)
x -= y # x = x - y
print(x)
x *= y # x = x * y
print(x)
x /= y # x = x / y
print(x)
x %= y # x = x % y
print(x)
x **= y # x = x ** y
print(x)
x //= y # x = x // y
print(x)
```
